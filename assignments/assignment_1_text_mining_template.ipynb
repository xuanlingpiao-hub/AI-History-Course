```
# 1) 데이터 불러오기
import pandas as pd, re
df = pd.read_csv('/mnt/data/AI-History-Course-2025/data/sample_historical_texts.csv')
df.head()
```

```
# 2) 전처리 & 토큰화 함수 작성 (필요시 형태소 분석기 교체 가능)
def simple_tokenize(text):
    return re.findall(r"[가-힣]{2,}", text)
df['tokens'] = df['text'].apply(simple_tokenize)
df[['title','tokens']].head()
```
