# Week 2 — 역사 텍스트 데이터 마이닝 입문

```
import pandas as pd
from collections import Counter
import re
df = pd.read_csv('/mnt/data/AI-History-Course-2025/data/sample_historical_texts.csv')
def tokenize_korean(text):
    tokens = re.findall(r"[가-힣]{2,}", text)
    return tokens
df['tokens'] = df['text'].apply(tokenize_korean)
# 간단 빈도
freq = Counter([t for tokens in df['tokens'] for t in tokens])
freq_df = pd.DataFrame(freq.items(), columns=['token','count']).sort_values('count', ascending=False)
freq_df.head(10)
```
