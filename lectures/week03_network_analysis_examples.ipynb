# Week 3 — 네트워크 분석 기초 (NetworkX)

```
import pandas as pd, networkx as nx
nodes = pd.read_csv('/mnt/data/AI-History-Course-2025/data/east_asia_network_nodes.csv')
edges = pd.read_csv('/mnt/data/AI-History-Course-2025/data/east_asia_network_edges.csv')
G = nx.from_pandas_edgelist(edges, source='source', target='target', edge_attr=True)
# 중심성
deg = nx.degree_centrality(G)
bet = nx.betweenness_centrality(G)
import pandas as pd
cent = pd.DataFrame({'id': list(deg.keys()), 'degree': list(deg.values())}).merge(
    pd.DataFrame({'id': list(bet.keys()), 'betweenness': list(bet.values())}), on='id')
cent.sort_values('degree', ascending=False).head()
```
